Analyze Air Pollution Trends Year wise

```python
df_grouped = df.groupby("year")["AQI"].mean()

df_grouped.plot(kind='line',marker='o' ,linestyle='-')
plt.grid()
plt.xlabel("Year")
plt.ylabel("AQI")
plt.title("Air Quality Trend Over Time")
plt.show()
plt.savefig('plt.png')
```

Analyze Air Pollution Trends Month Wise

```python
df['month']=df['Date'].dt.month

monthly_aqi=df.groupby('month')['AQI'].mean()
plt.plot(monthly_aqi, marker='o', color='blue')
plt.grid()
plt.xlabel("Month")
plt.ylabel("AQI")
plt.title("Monthly Variation in AQI")
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.show()
```

Identify Major Pollutants

```python
pollutants=df[["co", "no", "no2", "o3", "so2", "pm2_5", "pm10", "nh3"]].mean()
pollutant= pollutants.sort_values(ascending=False)

pollutant.plot(kind='bar',color='red')
plt.grid()
plt.xlabel("Pollutants")
plt.ylabel('Average Concentration')
plt.title('Average Concentration of Different Pollutants')

plt.show()
```

TOP 10 Polluted city

```python
polluted_city=df.groupby('City')['AQI'].mean()
polluted_city_sorted = polluted_city.sort_values(ascending=False)

polluted_city.plot(kind='bar',color='red')

plt.xlabel("Polluted city")
plt.ylabel('Avrage conceentration')
plt.title('Different polluted city')
plt.show()
```

```python

worst_month = df.groupby("month")["AQI"].mean()


df.groupby("month")["AQI"].mean().plot(kind="bar", color="red", alpha=0.7)

plt.xlabel("Month")
plt.ylabel("Average AQI")
plt.title("Average AQI Per Month")

plt.show()

```

Identify How Many Days Pollution Exceeded Safe Levels

```python

safe_limits = {"pm2_5": 25, "pm10": 50, "no2": 40, "so2": 20, "co": 10}


exceed = (df[["pm2_5", "pm10", "no2", "so2", "co"]] > pd.Series(safe_limits)).sum()
exceeded = exceed.sort_values(ascending=False)
exceeded.plot(kind="bar", color="orange")

plt.grid()
plt.xlabel("Pollutants")
plt.ylabel("Days Exceeding Safe Limit")
plt.title("Number of Days Pollutants Exceeded Safe Limits")
plt.show()
```

 Find Correlations Between Pollutants

```python

corr_matrix = df[["co", "no", "no2", "o3", "so2", "pm2_5", "pm10", "nh3"]].corr()

plt.figure(figsize=(8, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Between Pollutants")
plt.show()
```

```python

cleanest_days = df.nsmallest(10, "AQI")[["Date", "AQI"]]

most_polluted_days = df.nlargest(10, "AQI")[["Date", "AQI"]]
print('The top 10 most cleanest days')
print(" Cleanest Days:\n", cleanest_days)
print('The top 10 most polluted days')
print("\nMost Polluted Days:\n", most_polluted_days)

```
